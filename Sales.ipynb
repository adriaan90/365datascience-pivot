```python
import pandas as pd
```

```python
data = pd.read_excel("2_3_mapping-before.xlsx", sheet_name='Data',converters={'Period':str})
```

```python
data.shape
```

```python
data.dtypes
```

```python
data.isna().sum()
```

```python
data = data.dropna()
```

```python
data.isna().sum()
```

```python
from datetime import datetime
```

```python
data["Period_datetime"] = data["Period"].apply(lambda x: datetime.strptime(str(x), "%Y%m"))
```

```python
data.dtypes
```

```python
data["period_year"] = data["Period_datetime"].apply(lambda x: x.year)
```

```python
data["period_month"] = data["Period_datetime"].apply(lambda x: x.month)
```

```python
data.dtypes
```

```python
data.shape
```

```python
data.duplicated(subset=['Brand']).any()
```

```python
data.drop_duplicates(subset="Brand", inplace = True)
```

```python
data.shape
```

```python
pivot_data = data.pivot_table(index="Brand", columns = ["period_year"], values = ["Volume", "Gross Sales", "Discounts", "Net Sales", "Cost of Goods Sold", "Distribution", "Warehousing"], aggfunc = "sum", margins = True)
```
